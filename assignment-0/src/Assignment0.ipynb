```python
#Q4......................................
def cvt_2Gray(img):
  grayImg = np.zeros(img.shape,dtype=float)
  for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      grayImg[r][c]=img[r][c][0]*0.2989 +img[r][c][1]*0.587+img[r][c][2]*0.114
  return grayImg

img = cv2.imread("4.jpg")
grayImg=cvt_2Gray(img)

cv2.imwrite('gray.jpg',grayImg)
```
